```python
import pandas as pd
```

```python
houses = pd.read_csv(f"C:/Users/GODIAH/Downloads/houserecommender/synthetic_house_recommendation_data (1).csv")
```

```python
houses["words_in_house_keywords"] = houses["house_keywords"].str.split().str.len()
houses
```

```python
houses.loc[houses["words_in_house_keywords"].between(1,20),"house_keywords"]
```

```python
house_data = houses[houses["words_in_house_keywords"] >= 25]
house_data
```

```python
# Create a copy
house_data = house_data.copy()
```

```python
# Remove duplicate house keywords
import ast

def remove_duplicates(keyword_list):
    """
    Remove duplicate keywords from a list while preserving the original order.
    """
    return list(dict.fromkeys(keyword_list))

house_data.loc[:, 'house_keywords'] = house_data['house_keywords'].apply(
    lambda x: ast.literal_eval(x) if isinstance(x, str) else x
)

house_data.loc[:, 'house_keywords'] = house_data['house_keywords'].apply(remove_duplicates)
```

```python
house_data['house_keywords'] = house_data['house_keywords'].apply(str)
house_data["new_words_in_keywords"] = house_data["house_keywords"].str.split().str.len()
house_data
```

```python
duplicate_coordinates = house_data[house_data.duplicated(subset=["latitude", "longitude"], keep=False)]

if len(duplicate_coordinates) > 0:
    print(f"Found {len(duplicate_coordinates)} records with duplicate coordinates:")
    print(duplicate_coordinates)
else:
    print("No duplicate coordinates found in the dataset.")
```

```python
house_data_no_duplicates = house_data.drop_duplicates(subset=["latitude", "longitude"], keep='first')

records_removed = len(house_data) - len(house_data_no_duplicates)
print(f"\nRemoved {records_removed} duplicate coordinate records.")
print(f"Original dataset size: {len(house_data)}")
print(f"New dataset size: {len(house_data_no_duplicates)}")
```

```python
house_data_no_duplicates
```

```python
# Combine latitude and longitude
house_data_no_duplicates = house_data.drop_duplicates(subset=["latitude", "longitude"], keep='first').copy()
house_data_no_duplicates["co-ordinates"] = house_data_no_duplicates[["latitude","longitude"]].astype(str).agg(" ".join,axis=1)
house_data_no_duplicates
```

```python
# Create a unique identifier for each house
house_data_no_duplicates["tag"] = range(1, len(house_data_no_duplicates) + 1)
house_data_no_duplicates
```

```python
# Count records with less than 100 house keywords
count_less_than_100 = house_data_no_duplicates[house_data_no_duplicates["new_words_in_keywords"] < 100].shape[0]
print("Records with less than 100 keywords:", count_less_than_100)
```

```python
# Drop Records with more than 100 house keywords
houses_filtered = house_data_no_duplicates[house_data_no_duplicates["new_words_in_keywords"] <= 100]

# Optionally, check the shape of the filtered DataFrame
print("Records after dropping those with >100 keywords:", houses_filtered.shape[0])
```

```python
# Reset the tag index
houses_filtered = houses_filtered.reset_index(drop=True)

# Create a new "tag" column that auto-increments from 1
houses_filtered["tag"] = houses_filtered.index + 1
houses_filtered
```

```python
# Create new unique identifier
houses_filtered["house_tag"] = houses_filtered[["tag", "house_keywords"]].astype(str).agg(" ".join, axis=1)
houses_filtered
```

```python
{
    houses_filtered
    .drop(["Unnamed: 0","latitude","longitude","words_in_house_keywords"],axis=1)
    .to_csv("house_data_cleaned_v2.csv", index = False)
}
```
